```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv('Train_dataset.csv')
test_data = pd.read_csv('Test_dataset.csv')
```

```python
train_data.head()
```

```python
train_data.describe()
```

```python
train_feature = ['Age','Coma score','Diuresis','Platelets','HBB','d-dimer','Heart rate','HDL cholesterol','Charlson Index','Blood Glucose','FT/month','Infect_Prob']
test_feature = ['people_ID','Age','Coma score','Diuresis','Platelets','HBB','d-dimer','Heart rate','HDL cholesterol','Charlson Index','Blood Glucose','FT/month']
df = train_data[train_feature]
test_data = test_data[test_feature]
```

```python

df = df.dropna()
```

```python
df.info()
```

```python
df
```

```python
X = df.iloc[:,:-1]
```

```python
y = df.iloc[:,-1]
test_data
```

```python
X = X.astype('int')
y = y.astype('int')
```

```python
y
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X, y)
```

```python
log_pre = log_reg.predict(test_data.iloc[:,1:])
```

```python
test_data['Infect_Prob'] = log_pre
```
